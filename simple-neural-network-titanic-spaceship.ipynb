```python
#getting our data
training_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
#splitting our training data
X = training_data.loc[:, training_data.columns != "Transported"]
y = training_data["Transported"]
```

```python
#splitting further into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#splitting the categorical and numerical column names into different lists
numeric_columns = X_train.select_dtypes(np.float64).columns
categorical_columns = X_train.select_dtypes(object).columns
```

```python
from sklearn.preprocessing import OrdinalEncoder, StandardScaler #to encode categorical variables
from sklearn.impute import SimpleImputer #to impute numerical values
```

```python
# doing the ordinal encoding for categorical
ordinal = OrdinalEncoder()
X_train_ordinal = ordinal.fit_transform(X_train[categorical_columns])
```

```python
# imputing missing values
# no need to pass numeric cols seperate cause all should be numeric now
imputer = SimpleImputer()
X_train_ordinal_imputed = imputer.fit_transform(X_train_ordinal)
# X_train_imputed = imputer.transform(X_train_ordinal)
```

```python
# scaler
scaler = StandardScaler()
X_train_final = scaler.fit_transform(X_train_ordinal_imputed)
X_train_final.shape
```

```python
# creating our model
model = Sequential()
model.add(Dense(12,input_shape=(7,),activation='tanh'))
model.add(Dense(8, activation='relu'))
model.add(Dense (12, activation='tanh'))
model.add(Dense (7, activation='relu'))
model.add(Dense (1, activation='sigmoid'))
```

```python
# compliling the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
X_train = tf.convert_to_tensor(X_train_final)
y_train = tf.convert_to_tensor(y_train)
```

```python
# now to fit the model
# training occurs over epochs, each epoch is split into batches
model.fit(X_train,y_train,epochs=150, batch_size=10)
```
